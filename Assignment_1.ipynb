```python
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

# Load iris dataset
iris = datasets.load_iris()
data, labels = iris.data, iris.target

# Training-testing split
train_data, test_data, train_labels, test_labels = train_test_split(
    data, labels, train_size=0.8, test_size=0.2, random_state=12)

# Create and fit a KNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_data, train_labels)

# Predictions on training data
train_predictions = knn.predict(train_data)
print("Predictions from the classifier:")
print(train_predictions)
print("Target values:")
print(train_labels)
print(f"Training Accuracy: {accuracy_score(train_labels, train_predictions):.2f}")

# Predictions on test data
test_predictions = knn.predict(test_data)
print(f"Test Accuracy: {accuracy_score(test_labels, test_predictions):.2f}")

# Generate synthetic dataset
centers = [[2, 4], [6, 6], [1, 9]]
n_classes = len(centers)
syn_data, syn_labels = make_blobs(n_samples=150, centers=np.array(centers), random_state=1)

# Split synthetic data
train_syn_data, test_syn_data, train_syn_labels, test_syn_labels = train_test_split(
    syn_data, syn_labels, train_size=0.8, test_size=0.2, random_state=12)
```

```python
# Train KNN classifier on synthetic data
knn.fit(train_syn_data, train_syn_labels)

# Predictions on synthetic training data
syn_train_predictions = knn.predict(train_syn_data)
print("Predictions from classifier on synthetic training data:")
print(syn_train_predictions)
print("Actual Target Values:")
print(train_syn_labels)
print(f"Synthetic Training Accuracy: {accuracy_score(train_syn_labels, syn_train_predictions):.2f}")

# Predictions on synthetic test data
syn_test_predictions = knn.predict(test_syn_data)
print(f"Synthetic Test Accuracy: {accuracy_score(test_syn_labels, syn_test_predictions):.2f}")

# Plot classification results
plt.figure(figsize=(8,6))
plt.scatter(test_syn_data[:, 0], test_syn_data[:, 1], c=syn_test_predictions, cmap="viridis", edgecolor="k")
plt.title("KNN Classification Results on Test Data (k=3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Predicted Class")
plt.show()
```
